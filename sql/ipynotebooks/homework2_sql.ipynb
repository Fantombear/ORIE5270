```python
df5=pd.read_sql_query('select trading_20141201.id, substr(trading_20141201.a,1,9) from "trading_20141201","trading_20141202" where trading_20141201.id=trading_20141202.id and not trading_20141201.id=2',con=engine)
df5.head()
```

```python
df6=pd.read_sql_query('select count(trading_20141201.a),avg(trading_20141201.id),sum(trading_20141201.id) from "trading_20141201","trading_20141202" where trading_20141201.id=trading_20141202.id and not trading_20141201.id=2',con=engine)
df6
```

```python
df7=pd.read_sql_query('select * from trading_20141201 order by id desc',con=engine)
df7.head()
```

like pattern and wildcards

```python
df8=pd.read_sql_query("select id,a,b from trading_20141201 where a like '%%NA'",con=engine)
df8.head()
```

```python
df9=pd.read_sql_query("select id,a,b from trading_20141201 where a in ('093001173NA','112639662PA')",con=engine)
df9.head()
```
